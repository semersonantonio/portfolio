## Problema de negócio a ser resolvido:
 A empresa possui um valor aportado para investir, e quer saber qual será a estimativa de retorno do investimento nas diferentes interações com os clientes, seja no site ou no app, onde seria mais lucrativo estar investindo?

### Pré-processamento dos dados 

```python
dados.columns
```

```python
# Var. de entrada 
X = dados[['tempo_cadastro_cliente', 
           'numero_medio_cliques_por_sessao',
           'tempo_total_logado_app', 
           'tempo_total_logado_website']]
```

```python
# Var. de saída 
y = dados['valor_total_gasto']
```

```python
# Divisão em dados de treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
len(X_treino)
```

```python
len(X_teste)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_treino)
```

```python
X_treino = scaler.transform(X_treino)
```

```python
X_teste = scaler.transform(X_teste)
```

```python
X_treino[:5]
```

```python
X_teste[:5]
```

### Modelo 1 - Regressão Linear

```python
# Modelo
modelo_v1 = LinearRegression()
```

```python
# Treinamento
modelo_v1.fit(X_treino, y_treino)
```

```python
print('Coeficientes: \n', modelo_v1.coef_)
```

```python
# Coeficientes das variáveis preditoras
df_coef = pd.DataFrame(modelo_v1.coef_, X.columns, columns = ['Coeficiente'])
df_coef
```

#### Avaliação do Modelo

```python
# Previsões com dados de teste
pred_v1 = modelo_v1.predict(X_teste)
```

```python
# 10 primeiras prev.
pred_v1[:10]
```

```python
# Valor médio gasto pelos clientes
print('Valor máximo gasto pelos clientes é de:', dados['valor_total_gasto'].max())
print('Valor médio gasto pelos clientes é de:', dados['valor_total_gasto'].mean())
print('Valor mínimo gasto pelos clientes é de:', dados['valor_total_gasto'].min())
print('O MAE - Erro Médio Absoluto é de:', mean_absolute_error(y_teste, pred_v1))
```

```python
# MSE 
mean_squared_error(y_teste, pred_v1)
```

```python
# RMSE 
np.sqrt(mean_squared_error(y_teste, pred_v1))
```

```python
# Coeficiente R2
r2_score(y_teste, pred_v1)
```

```python
# Variância Explicada
explained_variance_score(y_teste, pred_v1)
```

### Modelo 2 com Regressão Ridge

```python
# Modelo
modelo_v2 = Ridge(alpha = 1.0)
```

```python
# Treinamento
modelo_v2.fit(X_treino, y_treino)
```

```python
print('Coeficientes: \n', modelo_v2.coef_)
```

```python
# Coeficientes das variáveis preditoras
df_coef = pd.DataFrame(modelo_v2.coef_, X.columns, columns = ['Coeficiente'])
df_coef
```

```python
# Previsões com dados de teste
pred_v2 = modelo_v2.predict(X_teste)
pred_v2[:10]
```

```python
# MAE
mean_absolute_error(y_teste, pred_v2)
```

```python
# MSE
mean_squared_error(y_teste, pred_v2)
```

```python
# RMSE
np.sqrt(mean_squared_error(y_teste, pred_v2))
```

```python
# Coeficiente R2
r2_score(y_teste, pred_v2)
```

```python
# Variância Explicada
explained_variance_score(y_teste, pred_v2)
```

### Modelo 3 - Regressão LASSO

```python
# Cria o modelo
modelo_v3 = Lasso(alpha = 1.0)
```

```python
# Treinamento
modelo_v3.fit(X_treino, y_treino)
```

```python
print('Coeficientes: \n', modelo_v3.coef_)
```

```python
# Coeficientes
df_coef = pd.DataFrame(modelo_v3.coef_, X.columns, columns = ['Coeficiente'])
df_coef
```

```python
# Previsões com dados de teste
pred_v3 = modelo_v3.predict(X_teste)
pred_v3[:10]
```

```python
# MAE
mean_absolute_error(y_teste, pred_v3)
```

```python
# MSE
mean_squared_error(y_teste, pred_v3)
```

```python
# RMSE
np.sqrt(mean_squared_error(y_teste, pred_v3))
```

```python
# Coeficiente R2
r2_score(y_teste, pred_v2)
```

```python
# Variância Explicada
explained_variance_score(y_teste, pred_v3)
```

## Considerações

```python
metrics = {
    'Modelo': ['Linear Regression', 'Ridge', 'Lasso'],
    'RMSE': [np.sqrt(mean_squared_error(y_teste, pred_v1)), np.sqrt(mean_squared_error(y_teste, pred_v2)), np.sqrt(mean_squared_error(y_teste, pred_v3))],
    'R²': [r2_score(y_teste, pred_v1), r2_score(y_teste, pred_v2), r2_score(y_teste, pred_v3)],
    'MAE': [mean_absolute_error(y_teste, pred_v1), mean_absolute_error(y_teste, pred_v2), mean_absolute_error(y_teste, pred_v3)]
}

metrics_df = pd.DataFrame(metrics)

# Gráfico de barras para comparar as métricas
metrics_df.set_index('Modelo').plot(kind='bar', figsize=(8, 4), color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Comparação de Métricas dos Modelos', fontsize=16)
plt.ylabel('Valor', fontsize=12)
plt.xlabel('Modelo', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

    A partir da construção e teste dos 3 modelos, em consideração a taxa de erro (RMSE), foi escolhido seguir para produção o modelo BenchMark (primeiro modelo), visto que apresenta uma alta performance e uma boa explicação sobre a variação entre os dados, além de ser o modelo mais simples, o que torna sua execução mais rápida e exige menor poder computacional e menores custos a empresa, pois não houveram diferenças significativas que sustentem a escolha dos modelos mais complexos, como o gráfico demonstra a seguir.     
    
    A partir do MAE podemos ver que nosso modelo escolhido preve em média R$7.76 de erro, o que é um valor muito baixo se compararmos a média de gasto por cliente de R$499,31.  
    
    Com o RMSE  podemos ver que nosso modelo preve em média R$9.74 reais de erro, o que também é um valor muito baixo se compararmos com o valor médio gasto por cliente de R$499,31. O R2 de 98% e RMSE de 9.74 demonstram uma alta eficiência do modelo, pois obteve uma alta precisão, conseguindo  explicar 98% da variância entre os dados.

```python
coef_df = pd.DataFrame(modelo_v1.coef_, X.columns, columns=['Coeficiente'])
coef_df.sort_values('Coeficiente', ascending=False).plot(kind='barh', figsize=(8, 4), color='skyblue')
plt.title('Coeficientes do Modelo de Regressão Linear', fontsize=16)
plt.xlabel('Coeficiente', fontsize=12)
plt.ylabel('Variáveis', fontsize=12)
plt.tight_layout()
plt.show()
```

    A partir das previsões feitas com o modelo escolhido podemos identificar que:

    O aumento de 1 unidade ao tempo total logado no web site está associado ao aumento de R$ 0.68 no valor total gasto por cliente por mês. O que indica que possivelmente, não será eficiente e com grandes retornos o investimento de mais recursos no web site.

    O aumento de 1 unidade ao tempo total logado na app está associado ao aumento de R$ 38.57 no valor total gasto por cliente no mês. Logo, sobre a dúvida principal entre investir no App ou Site, fica demonstrado a partir das previsões que a margem de retorno em um investimento será maior investindo no App perante ao site.

    O aumento de 1 unidade ao número médio de cliques por sessão está associado ao aumento de R$ 26.24 no valor total total gasto por cliente no mês. O que indica que o cliente ao interagir em mais clicks com App, possui uma maior probabilidade de aumentar seu valor total gasto, logo investir na melhoria do App para que o cliente tenha mais opções de interação, pode aumentar o seu valor total gasto.

    Porém, o maior insight gerado pelo nosso modelo, foi que o aumento de 1 unidade no tempo de cadastro do cliente está associado ao aumento de R$ 63.74 no valor total gasto por cliente no mês. Logo, clientes que possuem mais tempo fidelizados, são os que possuem um maior gasto mensal. Isso demonstra que segundo as probabilidades geradas pelo nosso modelo, dedica recursos e politicas para maior retenção e fidelização a médio e longo prazo dos clientes, tende a aumentar o valor gasto mensal por tais, logo aumentando a receita da empresa e diminuindo gastos desnecessários ou que atualmente não trariam significativos avanços as vendas, como por exemplo a baixa assosciação do uso do web site com o valor total gasto por cliente.
